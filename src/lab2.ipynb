## Part - 2


- Replacing missing values by the average or mode of the attribute (based on attribute types)

- Draw a histogram of the attribute before and after replacing missing values

```python
# counting the null values across all columns 
data.isna().sum() 
```

_We observe that the missing values are present in the columns with categorical values. In order to tackle that, we will replace the missing values with the category that has the highest frequency in that particular column_

```python
print(data['workclass'].value_counts())
fig = plt.figure(figsize = (14, 10))
data['workclass'].hist()
plt.title("Original Distribution for Col: Workclass")
```

We see in the column: `workclass`, the attribute `private` has the highest frequency. Hence, we will replace our missing value(s) with this attribute.

```python
mask = data['workclass'].isnull()
data['workclass'][mask] = ' Private'

```

```python
print(data['workclass'].value_counts())
fig = plt.figure(figsize = (14, 10))
data['workclass'].hist()
plt.title("Updated Distribution for Col: Workclass")
```

```python
print(data['occupation'].value_counts())
fig = plt.figure(figsize = (14, 10))
data['occupation'].hist(xrot =90, bins = 20)
plt.title("Original Distribution for Col: Occupation")
```

_replacing the missing values in the column: occupation with the most frequeny attribute:  Prof-specialty_

```python
data['occupation'].replace(to_replace = np.nan, value = ' Prof-specialty', inplace = True)
```

```python
print(data['occupation'].value_counts())
fig = plt.figure(figsize = (14, 10))
data['occupation'].hist(xrot =90, bins = 20)
plt.title("udpated Distribution for Col: Occupation")
```

```python
print(data['native_country'].value_counts())
fig = plt.figure(figsize = (14, 10))
data['native_country'].hist(xrot =90, bins = 30)
plt.title("Original Distribution for Col: Native Country")
```

```python
data['native_country'].replace(to_replace = np.nan, value = ' United-States', inplace = True)
```

```python
print(data['native_country'].value_counts())
fig = plt.figure(figsize = (14, 10))
data['native_country'].hist(xrot =90, bins = 30)
plt.title("udpated Distribution for Col: Native Country")
```
