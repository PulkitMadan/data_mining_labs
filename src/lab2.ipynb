```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
data = pd.read_csv("../data/adult_data_set/adult.data", header = None)
data.columns = ['age', 'workclass', 'fnlwgt', 'education', 'education_num', 'marital_status', 'occupation', 
               'relationship', 'race', 'sex', 'capital_gain', 'capital_loss', 'hours_per_week', 'native_country', 'income']
data.head()
```

```python
data.info()
```

### For continuous cols (numeric data) calculating, avg, standard deviation, min and max values

```python
from pandas.api.types import is_numeric_dtype
```

```python
for col in data.columns: 
    if is_numeric_dtype(data[col]): 
        print('%s:' % (col))
        print('\t Mean               = %.2f' % data[col].mean())
        print('\t Standard deviation = %.2f' % data[col].std())
        print('\t Minimum            = %.2f' % data[col].min())
        print('\t Maximum            = %.2f' % data[col].max())
        print()
```
